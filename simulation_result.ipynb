```python
import pickle

simulation_files = [f"Simulation{i}_result.pickle" for i in range(5)]

# Loop through each file and load data
for file in simulation_files:
    with open(file, "rb") as f:
        loaded_list = pickle.load(f)

    print(f"Results from {file}:")
    print("xi_loss:", loaded_list["xi_loss"])
    print("phi_loss:", loaded_list["phi_loss"])
    print("iteration number:" + str(len(loaded_list["xi_loss"])))

    print("-" * 50)
```
